```python
import pandas as pd
import requests
import json
import numpy as np
import re
```

```python
#Set year to read correct csv
year = '2019'

#Read in csv as df and add geo cols
df = pd.read_csv(f'EADA_{year}_filtered.csv')
df['lat'] = np.nan
df['lon'] = np.nan
df.head()
```

```python
#Create lists to hold coord
lat = []
lon = []
```

```python
#Adapted from code at https://www.mediawiki.org/wiki/API:Geosearch and by Ed Bowen
#Queries wikipedia for geostamp on each institution_name

s = requests.Session()

url = "https://en.wikipedia.org/w/api.php"

#Use iterrows to iterrate over df
for index, row in df.iterrows():
    institution_name = row['institution_name']
    
    params = {
    "action": "query",
    "format": "json",
    "titles": institution_name,
    "prop": "coordinates"
    }
    
    r = s.get(url=url, params=params)
    data = r.json()
    pages = data['query']['pages']
    
    #If a geostamp is returned, add the lat and lon to corresponding lists, else (in case of error), add nan
    try: 
        for k, v in pages.items():
            lat.append(str(v['coordinates'][0]['lat']))
            lon.append(str(v['coordinates'][0]['lon']))
    
    except: 
        lat.append(np.nan)
        lon.append(np.nan)
        
    finally:
        print(f'Row {index} done.')
```

```python
#Convert coord lists to pd series
lat_series = pd.Series(data = lat)
lon_series = pd.Series(data = lon)

#Add series as cols to df
df['lat'] = lat_series
df['lon'] = lon_series
```

```python
#Use redirects to fill out missing entries
#Adapted from code at https://www.mediawiki.org/wiki/API:Redirects

s = requests.Session()

url = "https://en.wikipedia.org/w/api.php"
    
for index, row in df.iterrows(): 
    if pd.isna(row['lat']):
        institution_name = row['institution_name']
    
        params = {
            "action": "query",
            "format": "json",
            "titles": institution_name,
            "prop": "coordinates",
            "redirects": True
        }

        r = s.get(url=url, params=params)
        data = r.json()

        new_query = data['query']['pages']
        
        #If redirect successfully results in a geotagged page, add the corrdinates to the df
        #Otherwise add the college and index to lists of problems
        try:
            for k in new_query:
                df.at[index, 'lat'] = new_query[k]['coordinates'][0]['lat']
                df.at[index, 'lon'] = new_query[k]['coordinates'][0]['lon']
            
        except: 
            pass
        
        finally:
            print(f'Row {index} done.')
        
```

```python
#Fix schools still without coordinates using first opensearch result
#Adapted from code at https://www.mediawiki.org/wiki/API:Opensearch

s = requests.Session()

url = "https://en.wikipedia.org/w/api.php"
    
for index, row in df.iterrows(): 
    if pd.isna(row['lat']):
        institution_name = row['institution_name']
 
        params = {
            "action": "opensearch",
            "namespace": "0",
            "search": institution_name,
            "limit": "1",
            "format": "json"
        }

        r = s.get(url=url, params=params)
        data = r.json()
        
        try:
            new_name = data[1][0]
            
            params = {
                "action": "query",
                "format": "json",
                "titles": new_name,
                "prop": "coordinates",
                "redirects": True
            }

            r = s.get(url=url, params=params)
            data = r.json()

            new_query = data['query']['pages']

            #If redirect successfully results in a geotagged page, add the corrdinates to the df
            #Otherwise add the college and index to lists of problems
            try:
                for k in new_query:
                    df.at[index, 'lat'] = new_query[k]['coordinates'][0]['lat']
                    df.at[index, 'lon'] = new_query[k]['coordinates'][0]['lon']

            except:
                pass
            
        except: 
            pass
        
        print(f'Row {index} done.')
```

```python
#Find coordinates for colleges with '-Main Campus' label by removing it and requerying wikipedia
s = requests.Session()

url = "https://en.wikipedia.org/w/api.php"
    
for index, row in df.iterrows(): 
    if pd.isna(row['lat']) and '-Main Campus' in row['institution_name']:
        new_name = re.sub('-Main Campus', '', row['institution_name'])

        try:            
            params = {
                "action": "query",
                "format": "json",
                "titles": new_name,
                "prop": "coordinates",
                "redirects": True
            }

            r = s.get(url=url, params=params)
            data = r.json()

            new_query = data['query']['pages']

            #If redirect successfully results in a geotagged page, add the corrdinates to the df
            #Otherwise add the college and index to lists of problems
            try:
                for k in new_query:
                    df.at[index, 'lat'] = new_query[k]['coordinates'][0]['lat']
                    df.at[index, 'lon'] = new_query[k]['coordinates'][0]['lon']

            except:
                pass
            
        except: 
            pass
        
        print(f'Row {index} done.')
```
